```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
P = lambda di: 100*((di/14)**0.45)
sieve_sizes = np.array([ 0.075, 0.15, 0.3, 0.6, 1.18, 2.36, 4.75, 9.5, 12.5, 14])
log_sieve_sizes = np.log10(sieve_sizes)
percent_passing = np.array([P(sz) for sz in sieve_sizes])
```

```python
mpl.rcParams['figure.dpi'] = 300
```

```python
plt.plot(sieve_sizes, percent_passing, "-o")
ax = plt.gca()
ax.set_xscale("log")
plt.xticks(rotation=70)
ax.set_xticks([0.075, 0.15, 0.3, 0.6, 1.18, 2.36, 4.75, 9.5, 14])
ax.get_xaxis().set_major_formatter(mpl.ticker.ScalarFormatter())
plt.title("Fuller Maximum Density (Maximum Size Aggregate: 14mm)")
plt.axvline(x=4.75, linestyle="--", color="red")
plt.xlabel("Grain Size (log scale)")
plt.ylabel("Percent Finer Than")
plt.show()
```
